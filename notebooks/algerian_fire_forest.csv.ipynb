```python
# import raw data

import pandas as pd

df = pd.read_csv("../data/algerian_forest_fires_raw.csv", header=0, skiprows=[0])
```

```python
# fix headers 

df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
# remove trailing spaces from column names

df.columns = df.columns.str.strip()
df
```

```python
# remove where classes is classes

df = df[df.Classes != "Classes"]
df.head()
```

```python
df = df[df.day != "Sidi-Bel Abbes Region Dataset"]
df.head()
```

```python
df = df[df.day != "day"]
df.head()
```

```python
df.day.value_counts()
```

```python
# fix data types

df.info()
```

```python
df.day = df.day.str.strip().astype(str).astype(int)
df.month = df.month.str.strip().astype(str).astype(int)
df.year = df.year.str.strip().astype(str).astype(int)
df.Temperature = df.Temperature.str.strip().astype(str).astype(int)
df.RH = df.RH.str.strip().astype(str).astype(int)
df.Ws = df.Ws.str.strip().astype(str).astype(int)

```

```python
df.info()
```

```python
# df.Rain = df.Rain.str.strip().astype(str).astype(float)
# df.DMC = df.DMC.str.strip().astype(str).astype(float)
# df.FFMC = df.FFMC.str.strip().astype(str).astype(float)
# df.ISI = df.ISI.str.strip().astype(str).astype(float)
# df.BUI = df.BUI.str.strip().astype(str).astype(float)
```

```python
df.info()
```

```python
df = df[df.DC != "14.6 9"]

df.DC = df.DC.str.strip().astype(str).astype(float)
```

```python
df.FWI = df.FWI.str.strip().astype(str).astype(float)
```

```python
df.info()
```

```python
df.shape
```

```python
df.Classes.value_counts()
```

```python
# trim the classes 

df.Classes = df.Classes.str.strip()
df.Classes.value_counts()
```

```python
# df.Classes = df.Classes.str.strip()
df.Classes.value_counts()
```

```python
# missing data

df.isnull().sum()
```

```python
# check the distribution of FWI and the Class, Fire vs Not Fire

df.groupby(["Classes"]).FWI.mean()

# higher values of FWI, higher chance of fire
```

```python
# save this data

df.to_csv("../data/algerian_forest_fires.csv", index = False)
```
